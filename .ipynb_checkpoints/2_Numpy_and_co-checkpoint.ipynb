# 1- Packages



Python has a large standard library, commonly cited as one of Python's greatest strengths, providing tools suited to many tasks. the official repository containing third-party software for Python, contains over 107,000 packages.

A *package* is a collection of *modules* i.e. groups of functions, classes, constants, types, etc.

## Built-in modules

To use a module, you have to *import* it using the command `import`.

```python
import math
```

You can now use it in the code by using its name as a prefix. 

```python
# The code calculates the cosine of 360° (a full circle), which equals 1.0.
x = math.cos(2 * math.pi)

print(x)
```

To explore the function and other content of the module/library:
 * Use the web documentation (e.g. for the `math` library <a href="https://docs.python.org/3/library/math.html">Doc for Python 3</a>)
 * Use the built-in`help`

```python
help(math)
```

```python
help(math.sqrt)
```

Using the name prefix can make the code obfuscated as it can get quite verbose (e.g. `scipy.optimize.minimize`) so Python provides simpler ways to import:
* `import name as nickname`: the prefix to call is now `nickname`

```python
import math as m

print(m.pi)
```

* `from name import function1,constant1` : `function1`  `constant1` can now be called directly. You can even import all contents with `from name import *` but this may be dangerous as names may conflict or override former ones, it is thus not advised except on user-generated modules.

```python
from math import e,log

print(log(e**4))
```

# 3- Linear Algebra 


Numpy comes with an efficient linear algebra module named `linalg` (see the [documentation](https://numpy.org/doc/stable/reference/routines.linalg.html) ). As in many languages, the more vectorized the operations are, the more efficient.

## Decompositions


* *QR:* `linalg.qr` Factor the matrix $A$ as $QR$, where $Q$ is orthonormal and $R$ is upper-triangular.
* *Cholesky:* `linalg.cholesky` Return the Cholesky decomposition, $L L^H$, of the square matrix $A$, where $L$ is lower-triangular. $A$ must be Hermitian and positive-definite. Only $L$ is actually returned.
* *SVD:* `linalg.svd`  Factors the matrix $A$ as $U  \text{diag}(s) V$, where $U$ and $V$ are unitary and $s$ is a 1-d array of $A$‘s singular values.



```python
A = np.random.randn(3,2)
```

```python
A
```

This code performs QR decomposition (factorization) on matrix A.


Why is it useful?

* Solving linear systems
* Least squares problems
* Computing eigenvalues
* Numerical stability in calculations

In simple terms: Splits matrix A into two special matrices - Q (with perpendicular unit-length columns) and R (triangular), such that multiplying them back gives you the original A.

```python
Q, R = np.linalg.qr(A)
print(A)
print(Q)
print(R)
```

```python
# This code checks if A equals Q×R (verifies the QR decomposition worked correctly).
np.allclose(A, np.dot(Q, R)) # check that A=QR
```

By default, $U$ and $V$ have the shapes $(M, M)$ and $(N, N)$  respectively if $A$ is $(M,N)$. If `full_matrices=False` is passed, the shapes are $(M, K)$ and $(K, N)$, respectively, where $K = min(M, N)$.

## Eigenvalues

`linalg.eig` compute the eigenvalues and right eigenvectors of a square array and is the main function (`linalg.eigvals` computes eigenvalues of a non-symmetric array, `linalg.eigh` returns eigenvalues and eigenvectors of a symmetric or Hermitian array).

Eigenvectors are special directions (vectors) that a matrix only stretches or shrinks, but doesn't rotate.
Mathematical definition:
For matrix A and eigenvector v:
A × v = λ × v
Where:

v = eigenvector (the special direction)
λ = eigenvalue (the stretch/shrink factor)

Simple explanation:
When you multiply a matrix by most vectors, the direction changes (rotation + stretching). But eigenvectors are special - they only get scaled (stretched or shrunk), keeping the same direction.

```python
A = np.array([[1, -1], [1, 1]])
print(A)
l, v = np.linalg.eig(A)
# print(l) - Prints the eigenvalues
# print(v) - Prints the eigenvectors as columns
print(l); print(v)
```

```python
print(v)
```

```python
# This code multiplies matrix A by the first eigenvector (first column of v).

A.dot(v[:,0])
```

We can check that $Ax= \lambda x$.

```python
# This code verifies that the first eigenvector calculation is correct by checking the fundamental eigenvalue equation.


np.allclose(A.dot(v[:,0]),l[0]*v[:,0])
```

```python
a = np.arange(9) - 4
B = a.reshape((3, 3))
```

In simple terms: Shows different ways to measure the "size" of a vector - standard length, sum of parts, biggest part, and count of non-zero parts.

```python
print(a)
print("none \t",np.linalg.norm(a)) # Euclidean norm (L2 norm) [DEFAULT]
print("2 \t",np.linalg.norm(a,ord=2)) # Euclidean norm (L2 norm) [DEFAULT]
print("1 \t",np.linalg.norm(a,ord=1)) # Manhattan norm (L1 norm)
print("inf \t",np.linalg.norm(a,ord=np.inf)) # Infinity norm (L∞ norm, max norm)
print("0 \t",np.linalg.norm(a,ord=0)) # L0 "norm" 
```

```python
print(B)
print("none \t",np.linalg.norm(B))
print("2 \t",np.linalg.norm(B,ord=2))
print("1 \t",np.linalg.norm(B,ord=1))
print("inf \t",np.linalg.norm(B,ord=np.inf))
print("fro \t",np.linalg.norm(B,ord='fro'))
```

Other useful function include the condition number `linalg.cond`, or rank `linalg.matrix_rank` .

```python
A = np.array([[1, 0, -1], [0, 1, 0], [1, 0, 1]]) # some matrix
I = np.eye(4) # identity
Def =  np.eye(4); Def[0,0]=0 # rank deficient , First diagonal changed to 0
```

In simple terms: Measures how "reliable" each matrix is for calculations. The identity matrix (I) has condition number 1 (perfect), while A's condition number reveals how sensitive it is to numerical errors.

```python
print(np.linalg.cond(A), np.linalg.cond(I))
```

 The condition number is infinity because the matrix is broken (rank deficient) - it has lost information and cannot be inverted, making it completely unreliable for solving equations.

```python
np.linalg.cond(Def)
```

This code calculates the rank of three different matrices.

Rank = number of linearly independent rows (or columns)

np.linalg.matrix_rank(Def) → 3

Def is 4×4 but has a zero row (first row is all zeros). 
Only 3 independent rows.
Rank deficient (rank = 3 < 4).
NOT invertible ✗

```python
print(np.linalg.matrix_rank(A), np.linalg.matrix_rank(I), np.linalg.matrix_rank(Def))
```
